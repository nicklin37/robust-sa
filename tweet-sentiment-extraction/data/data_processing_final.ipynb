```python
import nltk
import csv 
import pandas as pd
import numpy as np
import string
import re
import math
```

```python
# filename = "../data/adversarial_swap_train.csv"
# filename = "../data/train_final.csv"
filename = "../data/adversarial_typos_train.csv"

df = pd.read_csv(filename)
df = df.loc[df['sentiment'] != 'neutral']
df = df.dropna()

def calculate_class(x):
    if x == 'negative':
        return 0.0
    if x == 'positive':
        return 1.0
    return 0.0

df['classification'] = df['sentiment'].apply(lambda x : calculate_class(x))
```

```python
stopword = nltk.corpus.stopwords.words('english')
```

```python
def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text

def tokenization(text):
    text = text.strip()
    text = re.split('\W+', text)
    return text

def remove_stopwords(text):
    text = [word for word in text if word not in stopword]
    return text

df['clean_text'] = df['text'].apply(lambda x: remove_stopwords(tokenization(remove_punct(x).lower())))
df.head(10)
```

```python
wn = nltk.WordNetLemmatizer()

def lemmatizer(text):
    text = [wn.lemmatize(word) for word in text]
    return text

df['clean_text'] = df['clean_text'].apply(lambda x: lemmatizer(x))
df.head(10)
```

```python
df = df.loc[df['clean_text'].str.len() > 0]
```

```python
df['clean_text'] = df['clean_text'].apply(lambda x: ' '.join(x))
```
